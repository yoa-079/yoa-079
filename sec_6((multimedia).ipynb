```python
from skimage.transform import resize , rescale
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/img.jpg", cv2.IMREAD_GRAYSCALE)
rows,columns = img.shape

imge_resized =  resize(img, ( rows//5 , columns//6 ))
plt.imshow( img, cmap="gray")
plt.show

plt.imshow(imge_resized , cmap = "gray")
plt.show()
```

```python
from skimage.transform import resize , rescale
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/img.jpg", cv2.IMREAD_GRAYSCALE)

imge_rescaled_25 = rescale(img , 0.25)
imge_rescaled_50 = rescale(img , 0.5)
imge_rescaled_75 = rescale(img , 0.75)

plt.imshow( img, cmap='gray')
plt.show()
plt.imshow( imge_rescaled_25, cmap='gray')
plt.show()
plt.imshow( imge_rescaled_50, cmap='gray')
plt.show()
plt.imshow( imge_rescaled_75, cmap='gray')
plt.show()
```
